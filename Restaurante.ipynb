```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df_agrupado_por_dia = df.groupby('day',sort=False).agg({'total_bill':np.sum}).reset_index()
df_agrupado_por_dia
```

```python
px.bar(df_agrupado_por_dia, x='day', y='total_bill', title='Total de contas por final de semana')
```

```python
df_agrupado_por_dia_e_sexo = df.groupby(['day','sex'],sort=False).agg({'total_bill':np.sum}).reset_index()
df_agrupado_por_dia_e_sexo
```

```python
px.bar(df_agrupado_por_dia_e_sexo, x='day', y='total_bill', color='sex', barmode= 'group',title='Total de contas por final de semana e sexo')
```

```python
from datetime import time
px.scatter(df, x='total_bill', y= 'tip', color= 'day',hover_data= 'time')
```

```python
df_maiores_valores = df.nlargest(5,'total_bill')
df_maiores_valores
```

```python
df_menores_valores = df.nsmallest(5,'total_bill')
df_menores_valores
```

```python
df_maiores_valores_feminino = df[df['sex'] == 'Female'].nlargest(5, 'total_bill')
df_maiores_valores_feminino
```

```python
df_maiores_valores_masculino = df[df['sex'] == 'Male'].nlargest(5, 'total_bill')
df_maiores_valores_masculino
```
